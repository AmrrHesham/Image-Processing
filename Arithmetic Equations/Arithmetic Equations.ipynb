```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = "Image.jpg"
image = cv2.imread(image_path)

# Convert image to RGB for displaying
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert the image to float32 for arithmetic operations
image_float32 = image.astype(np.float32)
```

```python
# Addition
added_image1 = np.clip(image_float32 + 50, 0, 255).astype(np.uint8)
added_image2 = np.clip(image_float32 + 100, 0, 255).astype(np.uint8)
added_image3 = np.clip(image_float32 + 500, 0, 255).astype(np.uint8)


fig, axes = plt.subplots(1, 4, figsize=(15, 5))

axes[0].imshow(image_rgb)
axes[0].set_title("Original Image")

axes[1].imshow(cv2.cvtColor(added_image1, cv2.COLOR_BGR2RGB))
axes[1].set_title("Added 50")

axes[2].imshow(cv2.cvtColor(added_image2, cv2.COLOR_BGR2RGB))
axes[2].set_title("Added 100")

axes[3].imshow(cv2.cvtColor(added_image3, cv2.COLOR_BGR2RGB))
axes[3].set_title("Added 500")

plt.show()
```

```python
# Subtraction
subtracted_image1 = np.clip(image_float32 - 50, 0, 255).astype(np.uint8)
subtracted_image2 = np.clip(image_float32 - 100, 0, 255).astype(np.uint8)
subtracted_image3 = np.clip(image_float32 - 500, 0, 255).astype(np.uint8)


fig, axes = plt.subplots(1, 4, figsize=(15, 5))

axes[0].imshow(image_rgb)
axes[0].set_title("Original Image")

axes[1].imshow(cv2.cvtColor(subtracted_image1, cv2.COLOR_BGR2RGB))
axes[1].set_title("Subtracted 50")

axes[2].imshow(cv2.cvtColor(subtracted_image2, cv2.COLOR_BGR2RGB))
axes[2].set_title("Subtracted 100")

axes[3].imshow(cv2.cvtColor(subtracted_image3, cv2.COLOR_BGR2RGB))
axes[3].set_title("Subtracted 500")

plt.show()
```

```python
# Multiplication
multiplied_image1 = np.clip(image_float32 * 0.5, 0, 255).astype(np.uint8)
multiplied_image2 = np.clip(image_float32 * 1.5, 0, 255).astype(np.uint8)
multiplied_image3 = np.clip(image_float32 * 10, 0, 255).astype(np.uint8)

fig, axes = plt.subplots(1, 4, figsize=(15, 5))

axes[0].imshow(image_rgb)
axes[0].set_title("Original Image")

axes[1].imshow(cv2.cvtColor(multiplied_image1, cv2.COLOR_BGR2RGB))
axes[1].set_title("Multiplied 0.5")

axes[2].imshow(cv2.cvtColor(multiplied_image2, cv2.COLOR_BGR2RGB))
axes[2].set_title("Multiplied 1.5")

axes[3].imshow(cv2.cvtColor(multiplied_image3, cv2.COLOR_BGR2RGB))
axes[3].set_title("Multiplied 10")

plt.show()
```

```python
# Inverting the image
inverted_image = cv2.bitwise_not(image)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].imshow(image_rgb)
axes[0].set_title("Original Image")

axes[1].imshow(cv2.cvtColor(inverted_image, cv2.COLOR_BGR2RGB))
axes[1].set_title("Inverted Image")

plt.show()
```
